### create the thread

```python
thread = client.beta.threads.create()

# Create a message in the thread asking to create a narrative skeleton in JSON
message = client.beta.threads.messages.create(
    thread_id=thread.id,  # Use dot notation here instead of ['id']
    role="user",
    content="Generate a Little Red Riding Hood tale narrative skeleton, including logically ordered beats of narrative plot points, remixing Propp's 31 Narratemes and Campbell's 12 Stages. They should, in order, cover the basic three-part structure of Little Red Riding Hood (the journey, the encounter with the wolf, and the rescue). However, multiple narratemes/stages/motifs, up to ten, can be combined innovatively to make and enrich this basic three-part plot."
)

print(thread)
```

### create the "run" which includes a message and the final prompt (PROMPT A)

```python
thread_id='' #insert threrad id from above

run = client.beta.threads.runs.create(
    thread_id=thread_id,
    assistant_id='', #input your assistant id
    additional_instructions=""
)

print(run)
```

### run the thread

```python
run_info = client.beta.threads.runs.retrieve(
  thread_id=thread_id,
  run_id=run.id
)

print(run_info)
print("Run status:", run_info.status)
```
