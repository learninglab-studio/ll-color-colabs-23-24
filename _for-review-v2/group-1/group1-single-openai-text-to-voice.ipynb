```python
# @title Text To Speech
output_name = "speech.mp3" # name your output
voice_effect = "alloy" # can be alloy, echo, fable, onyx, nova, shimmer
input_text = "This is my favorite way to speak!" 

from openai import OpenAI
from google.colab import userdata
from IPython.display import Audio, display
client = OpenAI(api_key = userdata.get('OPENAI_API_KEY'))

speech_file_path = output_name
response = client.audio.speech.create(
  model="tts-1",
  voice=voice_effect,
  input=input_text
)

response.stream_to_file(speech_file_path)

display(Audio(speech_file_path, autoplay=True))
```
