```python
## Direct URL to online image

from google.colab import userdata
from openai import OpenAI
from IPython.display import Image

client = OpenAI(api_key=userdata.get('OPEN_API_KEY'))

Question = "What is in this image?" 
Image_url = "https://t4.ftcdn.net/jpg/00/97/58/97/360_F_97589769_t45CqXyzjz0KXwoBZT9PRaWGHRk5hQqQ.jpg" 

response = client.chat.completions.create(
  model="gpt-4-vision-preview",
  messages=[
    {
      "role": "user",
      "content": [
        {"type": "text", "text": Question},
        {
          "type": "image_url",
          "image_url": {
            "url": Image_url,
          },
        },
      ],
    }
  ],
  max_tokens=300,
)

display(Image(url=Image_url))
print(response.choices[0].message.content)
```

```python
import base64
import requests
from google.colab import files
from IPython.display import Image, display
from io import BytesIO

api_key = userdata.get('OPEN_API_KEY')

def encode_image(image_file):
    return base64.b64encode(image_file).decode('utf-8')

# Upload the image and get the file name
uploaded = files.upload()
file_name, image_data = next(iter(uploaded.items()))

# Encode the image file to base64
base64_image = encode_image(image_data)

# Question
Question = "Whatâ€™s in this image?" # @param {type:"string"}

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}

payload = {
    "model": "gpt-4-vision-preview",
    "messages": [
        {
            "role": "user",
            "content": [
                {
                    "type": "text",
                    "text": "What is in this image?"
                },
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base64_image}"
                    }
                }
            ]
        }
    ],
    "max_tokens": 300
}

response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

# Display the uploaded image and description
display(Image(data=image_data))
print(response.json()['choices'][0]['message']['content'])
```
