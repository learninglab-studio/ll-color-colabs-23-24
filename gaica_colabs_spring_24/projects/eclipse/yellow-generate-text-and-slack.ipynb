```python
# @title Text Generation
system_prompt = "You are a spunky, creative 5 year old who speaks in rhymes" # @param {type:"string"} indications for the system to behave
user_prompt = "Tell me a story about the origin of the solar eclipse" # @param {type:"string"}


from openai import OpenAI
from google.colab import userdata
client = OpenAI(api_key = userdata.get('OPENAI_API_KEY'))



response = client.chat.completions.create(
  model="gpt-3.5-turbo-0125",
  messages=[
    {"role": "system", "content": system_prompt},
    {"role": "user", "content": user_prompt}
  ]
)


print(response.choices[0].message.content)

output = response.choices[0].message.content
```

```python
# send to slack

import requests
import json
from google.colab import userdata
bot_token = userdata.get('SLACK_BOT_TOKEN')

# The channel ID you want to send the message to
channel_id = userdata.get('slack_channel_id')

# Headers for the API request
headers = {
    'Authorization': 'Bearer %s' % bot_token,
    'Content-Type': 'application/json'
}

# The message you want to send
message = {
  'channel': channel_id,
  'text': output
}

# Post the message to Slack
response = requests.post('https://slack.com/api/chat.postMessage',
                         headers=headers, json=message)


if response.status_code != 200:
    raise ValueError(
        'Request to slack returned an error %s, the response is:\n%s'
        % (response.status_code, response.text)
    )
```
