Upload the image that you want to change into "Files" on the left side bar of the Colab. Then hover the mouse on the uploaded image and click the 3 dots. Then click on "Copy Path" and paste the path into the Generate Image Variation code form. **Note that the image has to be in png format.**

```python
# @title Generate Image Variation

from openai import OpenAI
from google.colab import userdata
from IPython.display import Image, display

client = OpenAI(api_key = userdata.get('OPENAI_API_KEY'))
image_path = "/content/mix.png" # @param {type:"string"}

response = client.images.create_variation(
  image=open(image_path, "rb"),
  n=2,
  size="1024x1024"
)

image_url = response.data[0].url
print(f"Image URL: {image_url}")
display(Image(url=image_url, width=500))
```
